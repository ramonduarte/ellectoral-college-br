```python
df_fixed = df.iloc[:27,range(18)]
df_fixed.dels = df_fixed.dels.astype(int)

# TODO: lot of agg potential for this function to be further explored
# def count_delegates(df: pd.DataFrame, colname = "2018_1T_R") -> dict:
def count_delegates(df: pd.DataFrame, colname) -> dict:
    d = defaultdict(list)
    
    for uf in range(len(df)):  # rooting for a row-indexing
        # df[df.columns[X]] is the Xth column
#         print(df[colname][uf])
        if df[colname][uf] > .5:
            # TODO: delegates are integers
            # TODO: get rid of this horrendous workaround
            d[df[colname.replace("R", "V")][uf]].append(df.dels[uf])
        else:
            d["none"].append(int(df.dels[uf]))
    
    return d

def agg_delegates(d: dict) -> dict:
    agg = defaultdict(list)
    for k in d.keys():
        agg[k] = sum(d[k])
    return agg

# for k,v in agg_delegates(df_postcount).items():
#     print(k + ": ", v, "delegates")



# df_postcount = count_delegates(df_fixed, colname="2018_1T_R")
# df_postcount

df_fixed
```
